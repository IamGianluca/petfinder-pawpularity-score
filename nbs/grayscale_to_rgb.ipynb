The extra data we are using for pseudo labeling contains both RGB and grayscale images. To deal with that, we used this simple trick suggested by `ptrblck` ([link](https://discuss.pytorch.org/t/input-3-and-1-channel-input-to-the-network-in-pytorch/18519/2)).

```python
from PIL import Image
import numpy as np
```

When converted to an array or tensor, RGB images have shape (h, w, c), where c=3.

```python
img = Image.open('../data/extra/6231ed9b3-1.jpg').convert('RGB')
print(np.array(img).shape)
img
```

Grayscale images, however, have shape (h, w).

```python
img = Image.open('../data/extra/6231ed9b3-9.jpg')
print(np.array(img).shape)
img
```

Using `ptrblck`'s trick, however, we can convert grayscale images to RGB, which have (h, w, c) shape.

```python
img = Image.open('../data/extra/6231ed9b3-9.jpg').convert('RGB')
print(np.array(img).shape)
img
```
